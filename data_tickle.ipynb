```python
# Constantes de la physique
# ABSOLUMENT RECHARGER APRÈS AVOIR EXECUTÉ LES CASES D'IMPORT AU DESSUS

C_e = 1.602e-19        # Coulomb
kb = 1.38064852*1e-23  # Boltzman
m_Ca = 40.078*1.66054e-27 # masse Ca 40.078
m_GM = 1e6*1.66054e-27 # la masse de la GMol
eps0 = 8.854187*1e-12  # permittivité électrique du vide


r0 = 2.5e-3 # 2.5e-3   # rayon piège Ca+
d0 = (4e-3)/2            # longueur piège Ca+
Omega = 2.045e6*2*pi # 2.047e6
bk = 4 # nombre de barreaux par groupe (2 -> 4-pole , 4 -> 8-pole ...)

mkappa = 0.23          # écrantage piège réel GiantMol
wzLC = (2*pi*90806.9982303)**2
kappa_simion = m_Ca*d0**2*wzLC/(2*C_e)
print('%s = %f' % ('$\kappa_{simion}$',kappa_simion) )

zeta = kappa_simion*r0**2/d0**2
```

```python
Udc = 1
Urf = 42.43
omega_z_2 = 2*C_e*mkappa*Udc/m_Ca/d0**2
omega_z_2_simion = 2*C_e*kappa_simion*Udc/m_Ca/d0**2
omega_x_2 = 2*(C_e*Urf/m_Ca/Omega/r0**2)**2
omega_r_2 = omega_x_2 - 0.5*omega_z_2

qx = 4*C_e*Urf/(m_Ca*r0**2*Omega**2)
ax = 8*C_e*0/(m_Ca*r0**2*Omega**2)
az = 2*omega_z_2/Omega**2

print('alpha =',omega_z_2/omega_x_2)
print('qx =',qx)
print('ax =',ax)
print('az =',az)
print('f_z =',sqrt(omega_z_2)/2/pi)
print('f_x =',sqrt(omega_x_2)/2/pi)
print('f_r =',sqrt(omega_r_2)/2/pi)
```

```python
PM_all = [list(PM[0][:1+len(PM[0])//2])]
bonus = [0,PM[0][-1]-PM[1][1999//2],PM[0][-1]-PM[1][1999//2]+PM[1][-1]-PM[2][1999//2]]
for k in range(3):
    PM_all.append(PM[k][len(PM[k])//2:]+bonus[k])
# PM_all = [list(PM[0][:1+len(PM[k])//2]),PM_all]
PM_all = [item for sublist in PM_all for item in sublist]
# PM_all = [list(PM[0][:len(PM[k])//2]),PM_all]

print(len(PM[k]))
print(len(PM_all))
print(PM_all)
```

```python
testm_min = argrelextrema(array(PM_all),less,order=150)
testm_min = delete(testm_min,(0,1,2,4,5,9,10,11))

my_udc = 2.5
f_r = sqrt(omega_r_2)/2/pi
f_z = sqrt(omega_z_2_simion*my_udc)/2/pi

fig=figure('Total_plot',clear='True')
fig.set_size_inches(12, 7)
ax = fig.add_subplot(1, 1, 1)

# for k in range(3):
#     freq=linspace(100*k,200+100*k,1999)
#     plot(freq,PM[k],ls='--',alpha = 0.5)

freq=linspace(0,400,4000)

plt.plot(freq,PM_all,color='xkcd:azul',linewidth=1.5,alpha=1,ls='-')
plt.vlines(x=freq[testm_min],ymin=0,ymax=100000,linestyle='--')
plt.vlines(x=f_r/1000,ymin=0,ymax=120000,linestyle=':')
plt.vlines(x=f_z/1000,ymin=0,ymax=120000,linestyle=':')
# plt.vlines(x=130.5*2,ymin=0,ymax=80000,linestyle=':',color='red')

# ylim(-1500,100000)
# title('Tickle {}'.format(file_path[0][31:39]))
ax.set_xlabel('Frequency (kHz)')
ax.set_ylabel('Fluorescence (counts)')
ax.set_xlim(98,402)
# ax.set_ylim(0,120000)
# legend([file_path[0][40:60]])

# toto = ['$f_z$','$2f_z$','$f_x$','$f_y$']
toto = ['$f_z$','$2f_z$','$f_x$','$f_y$']

font = {'family': 'sans-serif',
        'color':  'xkcd:vermillion',
        'weight': 'normal',
        'size': 20,
        }
props = dict(facecolor='white', alpha=1)
ypos=[65000, 60000, 35500, 62000]
for k in range(len(testm_min)):
    plt.text(freq[testm_min[k]]+3.5,int(PM_all[testm_min[k]])-6000,'{} = {:.1f}'.format(toto[k],freq[testm_min[k]]),
             fontdict=font,rotation='90',bbox=props)

plt.text(f_r/1000+4,117000,r'$f_r^{{{}}}$={:.1f}'.format('sim',f_r/1000),
             fontdict=font,rotation='0',bbox=props)
plt.text(f_z/1000+4,117000,f'$f_z^{{sim}}$={f_z/1000:.1f}',
             fontdict=font,rotation='0',bbox=props)

ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=50))
ax.legend(['$U_{RF}=42.43$ V \n$V_{END}=2.5$ V \n$T_{int}=250$ ms \n$\delta f = 0.1$ kHz'],
          loc = 'lower left')
ax.grid()
plt.tight_layout()
savefig('tickle_demo.png')
```
