```python
# Constantes de la physique
# ABSOLUMENT RECHARGER APRÈS AVOIR EXECUTÉ LES CASES D'IMPORT AU DESSUS

C_e = 1.602e-19        # Coulomb
kb = 1.38064852*1e-23  # Boltzman
m_Ca = 40.078*1.66054e-27 # masse Ca 40.078
m_GM = 1e6*1.66054e-27 # la masse de la GMol
eps0 = 8.854187*1e-12  # permittivité électrique du vide


r0 = 2.5e-3 # 2.5e-3   # rayon piège Ca+
d0 = 4e-3/2            # longueur piège Ca+
Omega = 2.045e6*2*pi # 2.047e6
bk = 4 # nombre de barreaux par groupe (2 -> 4-pole , 4 -> 8-pole ...)

mkappa = 0.23          # écrantage piège réel GiantMol
wzLC = (2*pi*90806.9982303)**2
kappa_simion = m_Ca*d0**2*wzLC/(2*C_e)
print('%s = %f' % ('$\kappa_{simion}$',kappa_simion) )

zeta = kappa_simion*r0**2/d0**2
```

```python
Mk = 40
Ml = Mk

DL = 1e-2/Mk
DLz = 2*DL
Mp = int(5e-3/DL//2)

print(DL*1e3,'mm')

# Nz = 50

stgart = 1
end = 20
delta_reel = end - stgart
Ec = zeros((2*Mk,2*Ml,delta_reel,3))
Ed = zeros((2*Mk,2*Ml,delta_reel,3))

rho = 0.001
```

```python
print('start')

u_sur_l = 0

for x in [0]: # range(0)
    for y in range(-Ml,Ml):
        for z in range(stgart,end):
            print(x,y,z)
            for k in range(-Mk,+Mk):
                for l in range(-Ml,Ml):
#                     print(k,l)
                    u_sur_l = ((x*DL-k*DL)**2 + (y*DL-l*DL)**2 + (z*DLz)**2)**(-3/2)
                    Ec[x+Mk,y+Ml,z-stgart,0] += rho/(4*pi*eps0) * DL**2 * u_sur_l * (x*DL-k*DL)
                    Ec[x+Mk,y+Ml,z-stgart,1] += rho/(4*pi*eps0) * DL**2 * u_sur_l * (y*DL-l*DL)
                    Ec[x+Mk,y+Ml,z-stgart,2] += rho/(4*pi*eps0) * DL**2 * u_sur_l * (z*DLz)
                
            for k in range(-Mp,+Mp):
                Mq = int(sqrt(Mp**2-k**2))
                for l in range(-Mq,Mq):
                    u_sur_l = ((x*DL-k*DL)**2 + (y*DL-l*DL)**2 + (z*DLz)**2)**(-3/2)
                    Ed[x+Mk,y+Ml,z-stgart,0] += rho/(4*pi*eps0) * DL**2 * u_sur_l * (x*DL-k*DL)
                    Ed[x+Mk,y+Ml,z-stgart,1] += rho/(4*pi*eps0) * DL**2 * u_sur_l * (y*DL-l*DL)
                    Ed[x+Mk,y+Ml,z-stgart,2] += rho/(4*pi*eps0) * DL**2 * u_sur_l * (z*DLz)
```

```python
# chjeck cercle
fig = plt.figure('chjeck cercle',clear='True')
ax = fig.add_subplot(111)

for k in range(-Mp,+Mp):
    Mq = sqrt(Mp**2-k**2)
    ax.plot(k,Mq,'+')
    print(k**2+Mq**2)
ax.set_aspect('equal', adjustable='box')
ax.grid()
```
